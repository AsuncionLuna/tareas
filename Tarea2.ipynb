```python
# Problema 1
# ============
# Recorremos los números desde -30 hasta 50
for temperatura in range(-30, 51):

# Si la temperatura es menor que 0
    if temperatura < 0:
        print(temperatura, "Muy frío")

    # Si la temperatura está entre 0 y 24
    elif temperatura <= 24:
        print(temperatura, "Templado")

    # Si la temperatura es 25 o más
    else:
        print(temperatura, "Caluroso")
```

```python
# Problema 2
# ==========

for número in range(1, 101):
  # Si el número es divisible por 10
  if número % 10 == 0:
    print(número, "Bip Bip")

  # Si el número es divisible por 5
  elif número % 5 == 0:
    print(número, "Bip")
```

```python
# Problema 3
# ==========

puntaje1 = int(input("ingresa primer puntaje: "))

puntaje2 = int(input("ingresa segundo puntaje: "))

# Sumamos los puntajes
suma = puntaje1 + puntaje2

# Resultado
if suma < 50:
  print("Resultado: Nivel principiante")

elif suma < 80:
    print("Resultado: Nivel intermedio")

else:
    print("Resultado: Nivel pro")
```
